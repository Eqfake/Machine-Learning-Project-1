```python
# Import the sklearn library to access its datasets module
import sklearn.datasets
import numpy as np  # Import numpy for counting
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

#load the breast cancer dataset and Hand written digits data set
data_breast_cancer = sklearn.datasets.load_breast_cancer()
data_hand_written_digits = sklearn.datasets.load_digits()

#Process and display data involved with breast cancer data set
print(data_breast_cancer.feature_names)  # Print feature names
print(data_breast_cancer.target_names)    # Print target names
print(len(data_breast_cancer.target_names))  # Print number of classes
unique, counts = np.unique(data_breast_cancer.target, return_counts=True)  # Count occurrences of each class
class_counts = dict(zip(data_breast_cancer.target_names, counts))  # Create a dictionary of class names and their counts
print(class_counts)  # Print the number of samples in each class

#Process and display data involved with hand written digits
print(data_hand_written_digits.feature_names)  # Print feature names
print(data_hand_written_digits.target_names)    # Print target names
print(len(data_hand_written_digits.target_names))  # Print number of classes
unique, counts = np.unique(data_hand_written_digits.target, return_counts=True)  # Count occurrences of each class
class_counts2 = dict(zip(data_hand_written_digits.target_names, counts))  # Create a dictionary of class names and their counts
print(class_counts2)  # Print the number of samples in each class

# Access the feature data of the breast cancer data set
first_five_features_breast_cancer = data_breast_cancer.data[:, :5]  # Select the first 5 columns of the breast cancer dataset
mean_values_breast_cancer = np.mean(first_five_features_breast_cancer, axis=0)  # Calculate mean for each feature 
variance_values_breast_cancer = np.var(first_five_features_breast_cancer, axis=0)  # Calculate variance for each feature 

print("Mean of first five features (Breast Cancer):", mean_values_breast_cancer)
print("Variance of first five features (Breast Cancer):", variance_values_breast_cancer)

# Access the feature data of the handwritten digits dataset
first_five_features_hand_written = data_hand_written_digits.data[:, :5]  # Select the first 5 columns of the handwritten digits dataset
mean_values_hand_written = np.mean(first_five_features_hand_written, axis=0)  # Calculate mean for each feature 
variance_values_hand_written = np.var(first_five_features_hand_written, axis=0)  # Calculate variance for each feature 

print("Mean of first five features (Handwritten Digits):", mean_values_hand_written)
print("Variance of first five features (Handwritten Digits):", variance_values_hand_written)


# Count occurrences of each class for the breast cancer dataset
unique_breast_cancer, counts_breast_cancer = np.unique(data_breast_cancer.target, return_counts=True)

# Count occurrences of each class for the handwritten digits dataset
unique_hand_written, counts_hand_written = np.unique(data_hand_written_digits.target, return_counts=True)

# Plotting the distribution for the breast cancer dataset
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)  # Create a subplot for breast cancer
plt.bar(unique_breast_cancer, counts_breast_cancer, color='purple', alpha=0.7)
plt.title('Distribution of Labels (Breast Cancer)')
plt.xlabel('Labels')
plt.ylabel('Counts')
plt.xticks(unique_breast_cancer)  # Set x-ticks to be the unique labels

# Plotting the distribution for the handwritten digits dataset
plt.subplot(1, 2, 2)  # Create a subplot for handwritten digits
plt.bar(unique_hand_written, counts_hand_written, color='gold', alpha=0.7)
plt.title('Distribution of Labels (Handwritten Digits)')
plt.xlabel('Labels')
plt.ylabel('Counts')
plt.xticks(unique_hand_written)  # Set x-ticks to be the unique labels

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots

# Extract the first four features from the breast cancer dataset
first_four_features_breast_cancer = data_breast_cancer.data[:, :4]

# Extract the first four features from the handwritten digits dataset
first_four_features_hand_written = data_hand_written_digits.data[:, :4]

# Set up the figure for plotting
plt.figure(figsize=(12, 10))

# Plot histograms for the breast cancer dataset
for i in range(4):
    plt.subplot(2, 4, i + 1)  # Create a subplot for each feature
    plt.hist(first_four_features_breast_cancer[:, i], bins=30, color='purple', alpha=0.7)
    plt.title(f'Breast Cancer Feature {i + 1}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

# Plot histograms for the handwritten digits dataset
for i in range(4):
    plt.subplot(2, 4, i + 5)  # Create a subplot for each feature
    plt.hist(first_four_features_hand_written[:, i], bins=30, color='gold', alpha=0.7)
    plt.title(f'Handwritten Digits Feature {i + 1}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots

# Split the breast cancer dataset into training and test sets
X_breast_cancer = data_breast_cancer.data  # Features
y_breast_cancer = data_breast_cancer.target  # Labels

X_train_bc, X_test_bc, y_train_bc, y_test_bc = train_test_split(
    X_breast_cancer, y_breast_cancer, test_size=0.2, random_state=123
)

# Split the handwritten digits dataset into training and test sets
X_hand_written = data_hand_written_digits.data  # Features
y_hand_written = data_hand_written_digits.target  # Labels

X_train_hw, X_test_hw, y_train_hw, y_test_hw = train_test_split(
    X_hand_written, y_hand_written, test_size=0.2, random_state=123
)

# Print the shapes of the resulting datasets to verify the split
print("Breast Cancer Dataset:")
print("Training set shape:", X_train_bc.shape, y_train_bc.shape)
print("Test set shape:", X_test_bc.shape, y_test_bc.shape)

print("\nHandwritten Digits Dataset:")
print("Training set shape:", X_train_hw.shape, y_train_hw.shape)
print("Test set shape:", X_test_hw.shape, y_test_hw.shape)

# Standard Scaling for Breast Cancer dataset
standard_scaler_bc = StandardScaler()
X_train_bc_scaled = standard_scaler_bc.fit_transform(X_train_bc)  # Fit and transform the training data
X_test_bc_scaled = standard_scaler_bc.transform(X_test_bc)  # Transform the test data

# Min-Max Scaling for Breast Cancer dataset
min_max_scaler_bc = MinMaxScaler()
X_train_bc_minmax_scaled = min_max_scaler_bc.fit_transform(X_train_bc)  # Fit and transform the training data
X_test_bc_minmax_scaled = min_max_scaler_bc.transform(X_test_bc)  # Transform the test data

# Standard Scaling for Handwritten Digits dataset
standard_scaler_hw = StandardScaler()
X_train_hw_scaled = standard_scaler_hw.fit_transform(X_train_hw)  # Fit and transform the training data
X_test_hw_scaled = standard_scaler_hw.transform(X_test_hw)  # Transform the test data

# Min-Max Scaling for Handwritten Digits dataset
min_max_scaler_hw = MinMaxScaler()
X_train_hw_minmax_scaled = min_max_scaler_hw.fit_transform(X_train_hw)  # Fit and transform the training data
X_test_hw_minmax_scaled = min_max_scaler_hw.transform(X_test_hw)  # Transform the test data

# Print the shapes of the scaled datasets to verify
print("Breast Cancer Dataset - Standard Scaled:")
print("Training set shape:", X_train_bc_scaled.shape)
print("Test set shape:", X_test_bc_scaled.shape)

print("\nBreast Cancer Dataset - Min-Max Scaled:")
print("Training set shape:", X_train_bc_minmax_scaled.shape)
print("Test set shape:", X_test_bc_minmax_scaled.shape)

print("\nHandwritten Digits Dataset - Standard Scaled:")
print("Training set shape:", X_train_hw_scaled.shape)
print("Test set shape:", X_test_hw_scaled.shape)

print("\nHandwritten Digits Dataset - Min-Max Scaled:")
print("Training set shape:", X_train_hw_minmax_scaled.shape)
print("Test set shape:", X_test_hw_minmax_scaled.shape)

# Fit a logistic regression model for the Breast Cancer dataset
logistic_model_bc = LogisticRegression(max_iter=10000000, random_state=123)  # Create an instance of the model
logistic_model_bc.fit(X_train_bc_scaled, y_train_bc)  # Fit the model to the training data

# Fit a logistic regression model for the Handwritten Digits dataset
logistic_model_hw = LogisticRegression(max_iter=10000000, random_state=123)  # Create an instance of the model
logistic_model_hw.fit(X_train_hw_scaled, y_train_hw)  # Fit the model to the training data

# Print the coefficients of the fitted models
print("Breast Cancer Logistic Regression Coefficients:")
print(logistic_model_bc.coef_)

print("\nHandwritten Digits Logistic Regression Coefficients:")
print(logistic_model_hw.coef_)

# Predictions for Breast Cancer dataset
y_train_pred_bc = logistic_model_bc.predict(X_train_bc_scaled)  # Predictions on training set
y_test_pred_bc = logistic_model_bc.predict(X_test_bc_scaled)    # Predictions on test set

# Calculate accuracy, confusion matrix, and F1 score for Breast Cancer dataset
accuracy_train_bc = accuracy_score(y_train_bc, y_train_pred_bc)
accuracy_test_bc = accuracy_score(y_test_bc, y_test_pred_bc)
confusion_matrix_bc = confusion_matrix(y_test_bc, y_test_pred_bc)
f1_score_train_bc = f1_score(y_train_bc, y_train_pred_bc)
f1_score_test_bc = f1_score(y_test_bc, y_test_pred_bc)

# Print results for Breast Cancer dataset
print("Breast Cancer Dataset:")
print("Training Accuracy:", accuracy_train_bc)
print("Test Accuracy:", accuracy_test_bc)
print("Confusion Matrix:\n", confusion_matrix_bc)
print("Training F1 Score:", f1_score_train_bc)
print("Test F1 Score:", f1_score_test_bc)

# Predictions for Handwritten Digits dataset
y_train_pred_hw = logistic_model_hw.predict(X_train_hw_scaled)  # Predictions on training set
y_test_pred_hw = logistic_model_hw.predict(X_test_hw_scaled)    # Predictions on test set

# Calculate accuracy, confusion matrix, and F1 score for Handwritten Digits dataset
accuracy_train_hw = accuracy_score(y_train_hw, y_train_pred_hw)
accuracy_test_hw = accuracy_score(y_test_hw, y_test_pred_hw)
confusion_matrix_hw = confusion_matrix(y_test_hw, y_test_pred_hw)
f1_score_train_hw = f1_score(y_train_hw, y_train_pred_hw, average='weighted')  # Use 'weighted' for multi-class
f1_score_test_hw = f1_score(y_test_hw, y_test_pred_hw, average='weighted')      # Use 'weighted' for multi-class

# Print results for Handwritten Digits dataset
print("\nHandwritten Digits Dataset:")
print("Training Accuracy:", accuracy_train_hw)
print("Test Accuracy:", accuracy_test_hw)
print("Confusion Matrix:\n", confusion_matrix_hw)
print("Training F1 Score:", f1_score_train_hw)
print("Test F1 Score:", f1_score_test_hw)



# Step 1: Train and evaluate on original (unscaled) data for Breast Cancer dataset
logistic_model_bc_original = LogisticRegression(max_iter=10000000, random_state=123)
logistic_model_bc_original.fit(X_train_bc, y_train_bc)

# Predictions for original data
y_train_pred_bc_original = logistic_model_bc_original.predict(X_train_bc)
y_test_pred_bc_original = logistic_model_bc_original.predict(X_test_bc)

# Calculate accuracy, confusion matrix, and F1 score for original data
accuracy_train_bc_original = accuracy_score(y_train_bc, y_train_pred_bc_original)
accuracy_test_bc_original = accuracy_score(y_test_bc, y_test_pred_bc_original)
confusion_matrix_bc_original = confusion_matrix(y_test_bc, y_test_pred_bc_original)
f1_score_train_bc_original = f1_score(y_train_bc, y_train_pred_bc_original)
f1_score_test_bc_original = f1_score(y_test_bc, y_test_pred_bc_original)

# Print results for Breast Cancer dataset (original data)
print("Breast Cancer Dataset (Original Data):")
print("Training Accuracy:", accuracy_train_bc_original)
print("Test Accuracy:", accuracy_test_bc_original)
print("Confusion Matrix:\n", confusion_matrix_bc_original)
print("Training F1 Score:", f1_score_train_bc_original)
print("Test F1 Score:", f1_score_test_bc_original)

# Step 2: Use previously calculated results for scaled data
# Print results for Breast Cancer dataset (scaled data)
print("\nBreast Cancer Dataset (Scaled Data):")
print("Training Accuracy:", accuracy_train_bc)
print("Test Accuracy:", accuracy_test_bc)
print("Confusion Matrix:\n", confusion_matrix_bc)
print("Training F1 Score:", f1_score_train_bc)
print("Test F1 Score:", f1_score_test_bc)

# Repeat the same for Handwritten Digits dataset

# Step 1: Train and evaluate on original (unscaled) data for Handwritten Digits dataset
logistic_model_hw_original = LogisticRegression(max_iter=10000000, random_state=123)
logistic_model_hw_original.fit(X_train_hw, y_train_hw)

# Predictions for original data
y_train_pred_hw_original = logistic_model_hw_original.predict(X_train_hw)
y_test_pred_hw_original = logistic_model_hw_original.predict(X_test_hw)

# Calculate accuracy, confusion matrix, and F1 score for original data
accuracy_train_hw_original = accuracy_score(y_train_hw, y_train_pred_hw_original)
accuracy_test_hw_original = accuracy_score(y_test_hw, y_test_pred_hw_original)
confusion_matrix_hw_original = confusion_matrix(y_test_hw, y_test_pred_hw_original)
f1_score_train_hw_original = f1_score(y_train_hw, y_train_pred_hw_original, average='weighted')
f1_score_test_hw_original = f1_score(y_test_hw, y_test_pred_hw_original, average='weighted')

# Print results for Handwritten Digits dataset (original data)
print("\nHandwritten Digits Dataset (Original Data):")
print("Training Accuracy:", accuracy_train_hw_original)
print("Test Accuracy:", accuracy_test_hw_original)
print("Confusion Matrix:\n", confusion_matrix_hw_original)
print("Training F1 Score:", f1_score_train_hw_original)
print("Test F1 Score:", f1_score_test_hw_original)

# Step 2: Use previously calculated results for scaled data
# Print results for Handwritten Digits dataset (scaled data)
print("\nHandwritten Digits Dataset (Scaled Data):")
print("Training Accuracy:", accuracy_train_hw)
print("Test Accuracy:", accuracy_test_hw)
print("Confusion Matrix:\n", confusion_matrix_hw)
print("Training F1 Score:", f1_score_train_hw)
print("Test F1 Score:", f1_score_test_hw)
```
